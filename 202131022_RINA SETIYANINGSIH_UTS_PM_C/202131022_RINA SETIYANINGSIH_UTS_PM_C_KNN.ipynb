# Pemanggilan Dataset

```python
#Mengimpor pustaka pandas dengan memberikan alias pd
import pandas as pd

dataku = pd.read_csv('K03_counter_strike_2.csv')#Menggunakan fungsi read_csv dari pandas untuk membaca data
#dari file CSV. read_csv
dataku #Menyimpan hasil pembacaan data dari file CSV ke dalam variabel dataku
```

# Preprocessing Data

```python
from sklearn.preprocessing import LabelEncoder#Mengimpor kelas LabelEncoder dari modul preprocessing 
#dalam pustaka scikit-learn. digunakan untuk mengubah label kategorikal menjadi representasi numerik
from sklearn.model_selection import train_test_split#Mengimpor fungsi train_test_split dari modul 
#model_selection dalam scikit-learn.ntuk membagi data menjadi dua set: data latih (train set)
#dan data uji (test set). 
```

```python
encode = LabelEncoder()#Membuat objek LabelEncoder yang akan digunakan untuk mengubah nilai 
#kategorikal menjadi representasi numeri

for col in dataku:
    if dataku[col].dtype == 'object':
        dataku[col] = encode.fit_transform(dataku[col])
dataku
#Melakukan loop melalui setiap kolom pada DataFrame Jika tipe data
#kolom tersebut adalah 'object' (kategorikal), maka LabelEncoder 
#digunakan untuk mengubah nilai-nilai dalam kolom tersebut menjadi representasi numerik.
```

```python
X = dataku.drop('round_winner', axis=1)
Y = dataku['round_winner']
#Di sini, X adalah DataFrame yang berisi data fitur (variabel independen) dengan kolom 'round_winner' dihapus,
#dan Y adalah Series yang berisi data target (variabel dependen) yang hanya terdiri dari kolom 'round_winner'.

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=.2)
#X dan Y adalah data yang akan dibagi.
#test_size=0.2 menentukan bahwa 20% dari data akan dialokasikan sebagai data uji, 
#sedangkan 80% akan menjadi data latih
```

# Pembentukan Model

```python
from sklearn.neighbors import KNeighborsClassifier
#Ini mengimpor kelas KNeighborsClassifier dari modul neighbors dalam pustaka scikit-learn. 
#KNeighborsClassifier adalah implementasi K-Nearest Neighbors untuk tugas klasifikasi.

model = KNeighborsClassifier()#Membuat objek model KNeighborsClassifier. 
#Pada tahap ini, model diinisialisasi dengan parameter default.
model.fit(x_train, y_train)#x_train adalah data fitur latih, dan y_train adalah data target latih. 
#Model KNN akan "mempelajari" pola dalam data latih, sehingga dapat digunakan untuk melakukan prediksi pada data baru.
```

# Analisis Akurasi Model

```python
print("Akurasi Model Data : {}".format(model.score(x_test, y_test)))# digunakan untuk mencetak atau menampilkan akurasi dari model K-Nearest Neighbors pada data uji
```

# Pengujian Model

```python
predik = model.predict(x_test)
#Metode ini mengambil fitur-fitur dari data uji 
#dan mengembalikan prediksi kelas untuk setiap sampel.
#Hasil dari prediksi disimpan dalam variabel predik. 
#Sekarang, predik berisi array atau daftar nilai yang mewakili kelas prediksi dari model 
#untuk setiap sampel dalam data u
predik
```
